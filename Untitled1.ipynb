```python
result = list()
for i in range(2,5):
    name="result_"+str(i)+".json"
    with open(name,"r",encoding="gbk") as f:
        result= result+list(json.load(f))
result
```

```python
for idx,i in enumerate(result):
    i["id"]=idx
result
```

```python
def stopwordslist(filepath):
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]
    return stopwords
stopwords_list=stopwordslist("stopwords.txt")
```

```python
tmp= result[0]
import jieba
from collections import Counter 
# jieba.analyse.extract_tags(str(tmp["title"])
r = list()
abstract_list = [ i for i in list(jieba.cut(tmp["abstract"],cut_all = False)) if len(i)>1]
title_list = [ i for i in list(jieba.cut(tmp["title"],cut_all = False)) if len(i)>1]
author_list = "".join(tmp["author"]).split("，")
c = Counter([i for i in (abstract_list+title_list+author_list) if i not in stopwords_list])
```

```python
final_result=dict()
for text in result:
    tmp =text
    abstract_list = [ i for i in list(jieba.cut(tmp["abstract"],cut_all = False)) if len(i)>1]
    title_list = [ i for i in list(jieba.cut(tmp["title"],cut_all = False)) if len(i)>1]
    author_list = "".join(tmp["author"]).split("，")
    c = Counter([i for i in (abstract_list+title_list+author_list) if i not in stopwords_list])
    final_result[text['id']]=list(dict(c).keys())
final_result
```

```python
from pyspark import SparkConf,SparkContext
conf = SparkConf().setMaster("local[*]").setAppName("WordCount")
sc = SparkContext(conf=conf)
```

```python
data_rdd =sc.parallelize([ i for i in range(30)])
def map_func(x):
    l = list()
    for i in final_result[x]:
        l.append((i,x))
    return l
map_rdd = data_rdd.flatMap(map_func).groupByKey()
```

```python
res = map_rdd.collectAsMap()
```

```python
f = open("index.txt","w",encoding="gbk")
for key in res.keys():
#     print(key)
    f.write(str(key)+"\t")
    f.write("\t".join( [ str(i) for i in list(res[key])] ))
    f.write("\n")
```

```python
query = "SVM算法"
tags = jieba.cut(query)
tmp_result = list()
for tag in tags:
    tmp_result.append(list(res[tag]))

```
